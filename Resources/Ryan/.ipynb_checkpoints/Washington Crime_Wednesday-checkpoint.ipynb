# Histograms of Various Columns

### Create Function to Plot Histogram

```python
def plotHistogram(col, df):
    #Groupby operation on Target Column
    #Assign to temporarily DataFrame
    sub_df = df.groupby([col]).count()['CCN']
    sub_df = pd.DataFrame(sub_df)
    sub_df.sort_values('CCN', ascending=False, inplace=True)

    #Plot Histogram
    chart = sub_df.plot(kind="hist", color="yellowgreen", grid="True", figsize=(8,8), title=f"Histogram of {col}")
    chart.set_ylabel("Frequency of Offenses")
    plt.show()
```

### Create Histograms for Each Column

```python
cols = ['Report', 'Shift', 'Method', 'Offense', 'Block', 'XBlock',
       'YBlock', 'Ward', 'ANC', 'District', 'PSA', 'Neighborhood',
       'Block_Grou', 'Census_Trac', 'Voting_Precinct', 'Latitude', 'Longitude',
       'Bid', 'Start_Date', 'End_Date', 'Object_ID', 'Octo_Recor']
type(cols)
```

```python
for col in cols:
    plotHistogram(col,renamed_df)
```

# Bar Charts of Various Columns

### Create Function to Plot Bar Chart

```python
def plotBarChart(col, df):
    #Groupby operation on Target Column
    #Assign to temporarily DataFrame
    sub_df = df.groupby([col]).count()['CCN']
    sub_df = pd.DataFrame(sub_df)
    sub_df.sort_values('CCN', ascending=False, inplace=True)
    sub_df = sub_df.head(10)

    #Plot Histogram
    chart = sub_df.plot(kind="bar", color="indigo", grid="True", figsize=(8,8), title=f"Bar Chart of {col}")
    chart.set_ylabel("Number of Offenses")
    plt.show()
```

# Scatter Plots

### Function to Create Scatter Plot

```python
plt.figure(figsize=(8, 8))
plt.scatter(renamed_df.Census_Trac, renamed_df.XBlock, c="blueviolet", edgecolors="black",
           alpha=0.35)
plt.title("MPG vs displacement")
plt.xlabel("MPG")
plt.ylabel("displacement")
plt.show()
```

```python
def makeScatter(col1, col2, df):  
    plt.figure(figsize=(8, 8))
    plt.scatter(df[col1], df[col2], c="blueviolet", edgecolors="black", alpha=0.35)
    plt.title(f"{col1.title()} vs {col2.title()}")
    plt.xlabel(f"{col1.title()}")
    plt.ylabel(f"{col2.title()}")
    plt.tight_layout
    plt.show()
```

### Create List of Census Areas

```python
census_trac_list = sub_df.index.tolist()
census_trac_list
```

```python
def CreateCensusAreaLocation(area, df):
    
    #Filter for the Census Area
    mask = df['Census_Trac'] == area
    #Capture all the lat/long for the area
    temp_df = df[['Latitude', 'Longitude']].loc[mask]
    #Average the lat/long
    location = temp_df[['Latitude', 'Longitude']].mean()
    
    return location
```

```python
liz_list = CreateCensusAreaLocation(5800,renamed_df)
liz_list
```

```python
hin_list = []

for thing in census_trac_list:
    hin_list.append(CreateCensusAreaLocation(thing,renamed_df))
```

```python
hin_list
```

```python
num_of_crimes = sub_df['CCN'].to_list()
num_of_crimes
```

```python
locations = hin_list

figure_layout = {
    'width': '950px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}

fig = gmaps.figure(map_type="HYBRID", layout=figure_layout)
heat_layer = gmaps.heatmap_layer(locations, weights=num_of_crimes)

fig.add_layer(heat_layer)
fig
```
