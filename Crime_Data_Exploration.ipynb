# Plot Crimes by Offense over Time

### Groupby Start Date and Offense

```python
#Groupby 'Start_Date' by Month
crime_method_df = df_2017.groupby(["Start_Date","Offense"]).count()['CCN']
crime_method_df = pd.DataFrame(crime_method_df)
crime_method_df.reset_index(level=[0,1], inplace=True)
crime_method_df.head(3)
```

###  Rename Values for Offense

```python
crime_method_df.replace('THEFT F/AUTO', 'THEFT AUTO', inplace=True)
crime_method_df.replace('THEFT/OTHER', 'THEFT OTHER', inplace=True)
crime_method_df.replace('ASSAULT W/DANGEROUS WEAPON', 'ASSAULT WITH DANGEROUS WEAPON', inplace=True)
```

# Create Function to Plot Different Offense over Time

```python
def plotOffenseByMonth(offense, df):

    ###Filter by Offense
    mask = crime_method_df['Offense'] == offense
    crime_method_theft_auto = crime_method_df.loc[mask]
    
    
    ###Aggregate based on Month
    sub_df = crime_method_theft_auto.groupby(crime_method_theft_auto["Start_Date"].dt.month)['CCN'].count()
    sub_df = pd.DataFrame(sub_df)
    #Create Months Column
    sub_df["Months"] = months
    #Set Months as new Index
    sub_df.set_index('Months', inplace=True)
    
    
    ###Plot Graph
    x_axis = np.arange(len(sub_df))
    plt.figure(figsize=(20,8))

    #Colourmap legend (reverse colour_map: viridis_r)
    plot = plt.scatter(sub_df.CCN, sub_df.CCN, c = sub_df.CCN, cmap = 'viridis_r')
    plt.clf()
    plt.colorbar(plot)

    #Colourmap (reverse colour map:)
    colors = cm.viridis_r(sub_df.CCN / float(max(sub_df.CCN)))
    tick_locations = [value+0.4 for value in x_axis]

    plt.bar(sub_df.index, sub_df.CCN, color=colors, alpha=1.0, align="edge")
    plt.xticks(tick_locations, sub_df.index, rotation=0, fontsize=14)
    plt.ylabel("Number of Incidents", fontsize=14)
    #plt.xlabel("Day in the Month", fontsize=14)
    plt.title(f"Incidence of {offense} by Month (2017)", fontsize=20, fontweight="bold")
    plt.xlim(-0.25, len(sub_df)+0.25)
    plt.tight_layout()

    #Export Graph to File
    plt.savefig(f"Images/Offense_Incidence_of_{offense}_by_Month_2017.png")
    plt.show() 
```

```python
offenses = ['THEFT AUTO', 'BURGLARY', 'THEFT OTHER',
       'ASSAULT WITH DANGEROUS WEAPON', 'ROBBERY', 'MOTOR VEHICLE THEFT',
       'SEX ABUSE', 'HOMICIDE']
```

```python
for offense in offenses:
    plotOffenseByMonth(offense, crime_method_df)
```

# Bar Charts

### Create Function to Plot Bar Chart

```python
def matplotBarChart(col, df):
    #Groupby operation on Target Column
    #Assign to temporarily DataFrame
    #Sort & capture top 10
    sub_df = df.groupby([col]).count()['CCN']
    sub_df = pd.DataFrame(sub_df)
    sub_df.sort_values('CCN', ascending=False, inplace=True)
    sub_df = sub_df.head(10)

    #Plot Bar Chart
    plt.figure(figsize=(8, 8))
    plt.bar(sub_df.index, sub_df['CCN'], color='indigo', align="center")
    plt.xticks(rotation=90)
    plt.title(f"Washington Crime Statistics 2017 by {col}", fontsize=16, fontweight="bold")
    plt.xlabel(f"{col}")
    plt.ylabel("Number of Offences")
    axes = plt.gca()
    axes.yaxis.grid()
    plt.savefig(f'Images/Bar_Chart_{col}.png')
    plt.show
```

```python
def matplotBarChartHoriz(col, df):
    #Groupby operation on Target Column
    #Assign to temporarily DataFrame
    #Sort & capture top 10
    sub_df = df.groupby([col]).count()['CCN']
    sub_df = pd.DataFrame(sub_df)
    sub_df.sort_values('CCN', ascending=False, inplace=True)
    sub_df = sub_df.head(10)

    #Plot Bar Chart
    plt.figure(figsize=(8, 8))
    plt.bar(sub_df.index, sub_df['CCN'], color='indigo', align="center")
    plt.xticks(rotation=0)
    plt.title(f"Washington Crime Statistics 2017 by {col}", fontsize=16, fontweight="bold")
    plt.xlabel(f"{col}")
    plt.ylabel("Number of Offences")
    axes = plt.gca()
    axes.yaxis.grid()
    plt.savefig(f'Images/Bar_Chart_{col}.png')
    plt.show
```
